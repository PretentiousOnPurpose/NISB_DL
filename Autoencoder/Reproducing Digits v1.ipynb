```python
import tensorflow as tf
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
data = input_data.read_data_sets("data/")
```

```python
X = tf.placeholder(dtype=tf.float32 , shape=[None , 784])
```

```python
Z1 = tf.contrib.layers.fully_connected(inputs=X, num_outputs=392 , activation_fn=None)
```

```python
Z2 = tf.contrib.layers.fully_connected(inputs=Z1 ,num_outputs=196 , activation_fn=None)
```

```python
Z3 = tf.contrib.layers.fully_connected(inputs=X, num_outputs=392 , activation_fn=None)
```

```python
Z4 = tf.contrib.layers.fully_connected(inputs=X, num_outputs=784 , activation_fn=None)
```

```python
loss = tf.reduce_mean(tf.square(Z4 - X))
```

```python
opt = tf.train.AdagradOptimizer(learning_rate=0.5).minimize(loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for e in range(100):
    print("epoch: ", e)
    sess.run(opt , feed_dict={X:data.train.images.reshape((-1, 784))})
```

```python
# with 10 Epochs
res = sess.run(Z4, feed_dict={X:data.test.images[144].reshape(1 , 784)})
```

```python
plt.imshow(res.reshape((28, 28)))
```
